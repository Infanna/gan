```python
import numpy as np
from keras.preprocessing import image
from tensorflow.keras.preprocessing.image import load_img, img_to_array
import glob
import os

import tensorflow as tf
import matplotlib.pyplot as plt
```

```python
model = tf.keras.models.load_model('Model/generator_256px_300ep.h5')
```

```python
def generate_images(model, test_input, types):
    generate = model(test_input)
    fig = plt.figure(figsize=(8,8))
    row = len(test_input)

    for i, image in enumerate(generate):
        fig.add_subplot(row, 2, i*2+1)
        plt.imshow(test_input[i] / 2 + 0.5)
        plt.title("Originel Image (" + types + ")")
        plt.xticks([])
        plt.yticks([])

        fig.add_subplot(row, 2, i*2+2) 
        plt.imshow(image / 2 + 0.5)
        plt.title("Generate Image")
        plt.xticks([])
        plt.yticks([])
    
    plt.show()
```

```python
mylist1 = [f for f in glob.glob('./Img_Test_Model/128/*.png')]
img_list1 = [];

for i in mylist1 :
    img = load_img(i, color_mode = "rgb", target_size=(256,256,3))
    img = img_to_array(img)
    img = (img.astype(np.float32) / 127.5) - 1
    img_list1.append(img)
    
img_list1 = np.array(img_list1)
generate_images(model, img_list1, "128px")
```
